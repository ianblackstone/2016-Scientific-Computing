## Ballistic Deposition

Let us look at another random walk type scenario, this time with what is known as ballistic deposition.    Basically, we can consider it to be a model for how a film or other structure grows from depositing particles with an assumed trajectory pattern and assumed growth rules.

For instance, take the case where:
* particles are being sprayed down to a surface, and each particle falls randomly along a line
* for each particle, it can hit a part of the structure that is higher than nearby points, or go into a hole
* If the particle is in a column that is higher than its neighboring columns, it sticks to the top
* If the particle falls into a hole, it sticks to the first particle that it encounters at the same height

Of course, trajectories of incoming particles can be changed, "sticking" rules can be changed, the dimensionality can be changed, etc.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

#########################################

nparts=60000  #number of sites to fill
#create an array of index increments for use in populating sites
ny=500    #width of the grid or number of columns
nx=300     #height of the grid or number of rows

body=np.zeros((nx,ny))
#remember for array indexing first index is the row number, second the column

for i in range(nparts):
    column=int(np.random.rand()*ny)   #choose a random column
    
    curheight= np.where(body[:,column]==1)  
    if curheight[0].size == 0:
        height0=-1     #make height0 negative so no point added unless borderless
    else:
        height0=curheight[0][-1]   #check for 1's already in that column, get the highest row number
        
    if column > 0 and column < ny-1:       #not on the leftmost or rightmost column
        curheight= np.where(body[:,column-1]==1)
        if curheight[0].size == 0:
            heightm=-1
        else:
            heightm=curheight[0][-1]   #check for 1's already in that column, get the highest row number
        
        curheight= np.where(body[:,column+1]==1)
        if curheight[0].size == 0:
            heightp=-1
        else:
            heightp=curheight[0][-1]   #check for 1's already in that column, get the highest row number
    else:    #column is a border column
        if column == 0:
            heightm=-1
            curheight= np.where(body[:,column+1]==1)
            if curheight[0].size == 0:
                heightp=-1
            else:
                heightp=curheight[0][-1]   #check for 1's already in that column, get the highest row number
                 
        if column == ny-1:
            heightp=-1
            curheight= np.where(body[:,column-1]==1)
            if curheight[0].size == 0:
                heightm=-1
            else:
                heightm=curheight[0][-1]   #check for 1's already in that column, get the highest row number
                 

    if (height0 >= heightp) and (height0 >= heightm):
        body[height0+1,column]=1
    else:
        if heightm > heightp:
            body[heightm,column]=1
        else:
            body[heightp,column]=1


gsites=np.where(body==1)
gindex=np.transpose(gsites)  #change the order of gsites to create two element position indices

filled=np.where(body==1)
filled=np.transpose(filled)
#distances=np.arange(1,int(nx/2)-1)
distances=np.arange(1,125)

mass=np.zeros(distances.size)

#ctry=int(ny/2)
#ctrx=int(nx/4)

ctry=250
ctrx=125

for k in distances:
    for cur in filled:
        rdist=np.sqrt((ctrx-cur[0])**2 + (ctry-cur[1])**2)
        if rdist <= k:
            mass[k-1]+=1
plt.imshow(np.flipud(body))
plt.gray()
plt.show()        
```
